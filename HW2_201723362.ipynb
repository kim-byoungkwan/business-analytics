```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

# 1. 노드와 엣지를 직접 만들어서 네트워크를 만드는 방법.  

## 1.1 무방향성 네트워크와 가중치를 적용한 네트워크

```python
G_symmetric = nx.Graph()
G_symmetric.add_edge('아이유','g-dragon')
G_symmetric.add_edge('에픽하이','아이유')
G_symmetric.add_edge('에픽하이','오혁')
G_symmetric.add_edge('아이유','오혁')
G_symmetric.add_edge('HIGH4','아이유')
G_symmetric.add_edge('에픽하이','MINO')
G_symmetric.add_edge('에픽하이','사이먼도미닉')
G_symmetric.add_edge('에픽하이','더콰이엇')
G_symmetric.add_edge('에픽하이','수현')
G_symmetric.add_edge('MINO','사이먼도미닉')
G_symmetric.add_edge('MINO','더콰이엇')
G_symmetric.add_edge('사이먼 도미닉','더콰이엇')
```

```python
nx.draw_networkx(G_symmetric)
```

```python
nx.spring_layout(G_symmetric)
nx.draw_networkx(G_symmetric)
```

```python
G_weighted = nx.Graph()
G_weighted.add_edge('아이유','g-dragon',weight=25)
G_weighted.add_edge('에픽하이','아이유',weight=8)
G_weighted.add_edge('에픽하이','오혁',weight=11)
G_weighted.add_edge('아이유','오혁',weight=1)
G_weighted.add_edge('HIGH4','아이유',weight=4)
G_weighted.add_edge('에픽하이','MINO',weight=7)
G_weighted.add_edge('에픽하이','사이먼도미닉',weight=10)
G_weighted.add_edge('에픽하이','더콰이엇',weight=9)
G_weighted.add_edge('에픽하이','수현',weight=12)
G_weighted.add_edge('MINO','사이먼도미닉',weight=14)
G_weighted.add_edge('MINO','더콰이엇',weight=5)
G_weighted.add_edge('사이먼 도미닉','더콰이엇',weight=2)
```

```python
nx.draw_networkx(G_weighted)
```

## 1.2 양방향성 네트워크에 가중치를 적용

```python
G_digraph = nx.DiGraph()
G_digraph.add_edge('아이유','g-dragon',weight=25)
G_digraph.add_edge('에픽하이','아이유',weight=8)
G_digraph.add_edge('에픽하이','오혁',weight=11)
G_digraph.add_edge('아이유','오혁',weight=1)
G_digraph.add_edge('HIGH4','아이유',weight=4)
G_digraph.add_edge('에픽하이','MINO',weight=7)
G_digraph.add_edge('에픽하이','사이먼도미닉',weight=10)
G_digraph.add_edge('에픽하이','더콰이엇',weight=9)
G_digraph.add_edge('에픽하이','수현',weight=12)
G_digraph.add_edge('MINO','사이먼도미닉',weight=14)
G_digraph.add_edge('MINO','더콰이엇',weight=5)
G_digraph.add_edge('사이먼 도미닉','더콰이엇',weight=2)
```

```python
nx.draw_networkx(G_digraph)
```

# 2. csv 파일을 직접 불러와서 네트워크를 만드는 방법

```python
featuring = pd.read_csv('C:/Users/kwan/Desktop/capstone/personel_assignment/featuring.csv')
```

```python
g = nx.from_pandas_edgelist(featuring,'from',"to",create_using = nx.DiGraph())
nx.draw(g,with_labels=True)
```
